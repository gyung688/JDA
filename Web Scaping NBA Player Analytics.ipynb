```python
import requests
from bs4 import BeautifulSoup

link = "https://www.basketball-reference.com/"
data = requests.get(link)
data
```

```python
html_content = BeautifulSoup(data.text, "html.parser")
```

```python
table = {"Team":[], "Wins":[], "Losses":[], "Win/Loss %":[], "Games Behind":[]}
for i in html_content.find(id="mini-fib").find("tbody").find_all("tr"):
    table["Team"].append(i.find('a').text)
    row_data = i.find_all("td")
    table["Wins"].append(row_data[0].text)
    table["Losses"].append(row_data[1].text)
    table["Win/Loss %"].append(row_data[2].text)
    table["Games Behind"].append(row_data[3].text)
```

```python
import pandas as pd

nbaTable = pd.DataFrame(table)
nbaTable
```
